```python
noise_img=[]
```

```python
for i in range(len(images_list)):

  gausBlur = cv2.GaussianBlur(images_list[i], (5,5),0)
  noise_img.append(gausBlur)
```

```python
width = int(noise_img[0].shape[1]/3.5)
height = int(noise_img[0].shape[0]/3.5)
dim = (width, height)
```

```python
#resized = cv2.resize(noise_img[0], dim, interpolation = cv2.INTER_AREA)
```

```python
degraded_img=[]
```

```python
for i in range(len(noise_img)):

  resized = cv2.resize(noise_img[i], dim, interpolation = cv2.INTER_AREA)
  degraded_img.append(resized)
```

```python
for i in range(len(degraded_img)):
  path = "/content/drive/MyDrive/Degraded Images/"+str(i+1)+".png"
  cv2.imwrite(path, degraded_img[i])
```
